```python
import pandas as pd
import numpy as np
import csv
from bs4 import BeautifulSoup
import requests
import re
import string
url = "http://www.fightmetric.com/statistics/fighters?char="
```

```python
l=[]
def getUFCdata(url):
    with open('fightmetricclean.csv', 'w') as csvfile:
        writer=csv.writer(csvfile, delimiter=',')
        writer.writerow(["First", "Last", "Nickname", "Ht", "Wt", "Reach", "Stance", "Win", "Loss", "Draw", "Belt"])
        for char in string.ascii_lowercase:
            r = requests.get(url+char+"&page=all")
            html = BeautifulSoup(r.text, 'html.parser')
            fighters = list(html.find_all("tr", class_="b-statistics__table-row"))
            for fighter in fighters:
                row = list(fighter.find_all("td", class_="b-statistics__table-col"))
                link= fighter.find('a',class_="b-link b-link_style_black")
                l.append(link)
                try:
                    writer.writerow([ x.get_text().strip() for x in row])
                except:
                    print("missing data")
getUFCdata(url)
```

```python
fighterdata = pd.read_csv('fightmetricclean.csv')
fighterdata
```

```python
l = list(dict.fromkeys(l))
l.remove(None)
l
```

```python
fighterdata.info()
```

```python
links =[]
for i in l:
    links.append(i.get('href'))
```

```python
fighterdata['Links']=links
```

```python
fighterdata
```

```python
SLpM =[]
StrAcc=[]
SApM=[]
Strdef=[]
TDavg=[] 
TDAcc=[]
TDDef=[]
SubAvg=[]
for i in fighterdata['Links']:
    r = requests.get(i)
    html = BeautifulSoup(r.text, 'html.parser')
    stats=list(html.find_all("li",class_="b-list__box-list-item b-list__box-list-item_type_block"))
    SLpM.append(stats[5].text)
    StrAcc.append(stats[6].text)
    SApM.append(stats[7].text)
    Strdef.append(stats[8].text)
    TDavg.append(stats[10].text)
    TDAcc.append(stats[11].text)
    TDDef.append(stats[12].text)
    SubAvg.append(stats[13].text)
```

```python
fighterdata['sigstrikes_landed_permin'] = SLpM
fighterdata['strike_acc'] = StrAcc
fighterdata['sigstrikes_abs_permin'] = SApM
fighterdata['sigstrikes_def']= Strdef
fighterdata['takedown_avg_per15min']= TDavg
fighterdata['takedown_acc']= TDAcc
fighterdata['takedown_def']=TDDef
fighterdata['avg_sub_per15min']= SubAvg
```

```python
fighterdata['sigstrikes_landed_permin']= [re.findall('\d*\.?\d+',s) for s in fighterdata['sigstrikes_landed_permin'] ]
fighterdata['strike_acc'] = [re.findall('\d*\.?\d+',s) for s in fighterdata['strike_acc'] ]
fighterdata['sigstrikes_abs_permin'] = [re.findall('\d*\.?\d+',s) for s in fighterdata['sigstrikes_abs_permin']]
fighterdata['sigstrikes_def']= [re.findall('\d*\.?\d+',s) for s in fighterdata['sigstrikes_def' ]]
fighterdata['takedown_avg_per15min']= [re.findall('\d*\.?\d+',s) for s in fighterdata['takedown_avg_per15min']]
fighterdata['takedown_acc']= [re.findall('\d*\.?\d+',s) for s in fighterdata['takedown_acc'] ]
fighterdata['takedown_def']=[re.findall('\d*\.?\d+',s) for s in fighterdata['takedown_def'] ]
fighterdata['avg_sub_per15min']= [re.findall('\d*\.?\d+',s) for s in fighterdata['avg_sub_per15min'] ]
```
