```python
import pandas as pd
import numpy as np
import csv
from bs4 import BeautifulSoup
import requests
import re
import string
url = "http://www.espn.com/mma/fighters?search="
```

```python
name=[]
country=[]
link=[]
for char in string.ascii_lowercase:
    r = requests.get(url+char)
    html = BeautifulSoup(r.text, 'html.parser')
    links = list(html.find_all("td"))[3:]
    for i in range(int(len(links)/2)):
        name.append(links[i*2].text)
        country.append(links[(2*i)+1].text)
    for j in links:
        link.append(j.find('a')) 
        
                        
    
```

```python
l = list(dict.fromkeys(link))
l.remove(None)
l
```

```python
site=[]
for i in l:
    site.append(i.get('href'))
len(site)
```

```python
first=[]
last=[]
for i in name:
    first.append(i.split(',')[1].strip())
    last.append(i.split(',')[0].strip())
```

```python
espn=pd.DataFrame()
espn['First']=first
espn['Last']=last
espn['Country']=country
url='https://www.espn.com'
s=[]
for i in site:
    s.append(url+i)
espn['link']=s
```
